```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing import sequence
from tensorflow.keras.models import Sequential  ##used for any NN

from tensorflow.keras.layers import Embedding,SimpleRNN,Dense
```

```python
max_features=10000

(x_train,y_train),(x_test,y_test)=imdb.load_data(num_words=max_features)

print(f'Training data shape: {x_train.shape}, Training labels shape: {y_train.shape}')
print(f'Testing data shape: {x_test.shape}, Testing labels shape: {y_test.shape}')

```

```python
##inspect sample review

print(f'sample review : {x_train[0]}')  ##this is one of the sentence
print(f'sample lables: {y_train[0]}')
```

```python
word_index=imdb.get_word_index()
reverse_word_index={value:key for key,value in word_index.items()}

reverse_word_index 
```

```python
#whole review sentence
for ind in x_train[0]:
    print(reverse_word_index.get(ind-3,'?'),end=' ')
```

```python
max_len=500
x_train=sequence.pad_sequences(x_train,padding='pre',maxlen=max_len)
x_test=sequence.pad_sequences(x_test,padding='pre',maxlen=max_len)

print(x_train[0])
```

```python
#TRAINING RNN
model=Sequential()
model.add(Embedding(max_features,128,input_length=max_len))  ##embedding layer
model.add(SimpleRNN(128,activation='tanh'))  ##RNN layer
model.add(Dense(1,activation='sigmoid'))  ##output layer
model.build(input_shape=(None, max_len))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
##create instance of earlystopping callback

from tensorflow.keras.callbacks import EarlyStopping

early_stop=EarlyStopping(monitor='val_loss',patience=5,restore_best_weights=True)

early_stop
```

```python
model_history=model.fit(
    x_train,y_train,
    epochs=10,
    batch_size=32,
    validation_split=0.2,
    callbacks=[early_stop]
)
```

```python
model.save('imdbRNN.h5')
```
