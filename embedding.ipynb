```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
print("hello")
```

```python
sentence = ['I love my dog',
            'I love my cat',
            'You love my dog',
            'Do you think my dog is amazing?',
            'i am the smartest person in the class'
            
]
```

```python
vocab_size = 10000
```

```python
one_hot_rep=[one_hot(words,vocab_size)for words in sentence]
one_hot_rep
```

```python
##making the embeddings

from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential

import numpy as np
```

```python
max_sent_len=10

embedded_docs=pad_sequences(one_hot_rep,padding='pre',maxlen=max_sent_len)
embedded_docs
```

```python
#feature representation

dimension=10
```

```python
model=Sequential()
model.add(Embedding(vocab_size,dimension,input_length=max_sent_len))
model.compile('adam','mse')
```

```python

model.summary()
```

```python
predicted_doc=model.predict(embedded_docs)
```

```python
print("sentence before embedding: " ,embedded_docs[0])
print("sentence after embedding: ",predicted_doc[0][0])
```
